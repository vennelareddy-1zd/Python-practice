```python
import pandas as pd
uber_data=pd.read_csv("Uber_Drives_2016.csv")  #load uber.csv dataset
uber_data
```

```python
print(uber_data.shape) #display number of rows and columns
```

```python
uber_data['CATEGORY*'].unique() #show unique categories in CATEGORY* column
```

```python
null_rows=uber_data[uber_data['PURPOSE*'].isnull()] #find how many null values are there in PURPOSE* column
print(null_rows)
```

```python
uber_data=uber_data.rename(columns={'START_DATE*':'START_DATE','END_DATE*':'END_DATE','CATEGORY*':'CATEGORY','START*':'START','STOP*':'STOP','MILES*':'MILES','PURPOSE*':'PURPOSE'})
uber_data
```

```python
# ASSIGNMENT-2
```

```python
rides=uber_data[uber_data['CATEGORY']=='Business'] #display rides where CATEGORY=='Business'
rides
```

```python
top5=uber_data.sort_values(by='MILES',ascending=False) #show top5 rides with longest distance(miles)
top5.head(5)
```

```python
uber_data['PURPOSE']=uber_data['PURPOSE'].fillna('Not Specified') #replace all missing purpose values with 'not specified'
uber_data
```

```python
uber_data['START_DATE']=pd.to_datetime(uber_data['START_DATE'],errors='coerce') #create a new column tripduration using enddate-startdate
uber_data['END_DATE']=pd.to_datetime(uber_data['END_DATE'],errors='coerce')
uber_data['TRIP_DURATION']=uber_data['END_DATE']-uber_data['START_DATE']
uber_data
```

```python
sorttrips=uber_data.sort_values(by='MILES',ascending=False) #sort by distance in descending order
sorttrips
```
